Podemos enchular la grafica, es decir, le damos dimensiones, le ponemos etiquetas, le cambiamos el color, le ponemos nombre a los ejes

```python
fig, ax = plt.subplots(figsize=(10,5),dpi=100)
ax.plot(df['CO'], color='purple',lw=0.2);
ax.tick_params(labelsize=14)
ax.set_xlabel('Tiempo', fontdict = {'fontsize':14 })
ax.set_ylabel('CO ($\mu$g/$m^3$)', fontdict = {'fontsize':14 });
#ax.grid(True)
```

Grafica de otro contaminante

```python
fig, ax = plt.subplots(figsize=(10,5),dpi=100)
ax.plot(df['PM25'], color='blue',lw=0.4);
ax.tick_params(labelsize=14)
ax.set_xlabel('Tiempo', fontdict = {'fontsize':14 })
ax.set_ylabel('PM25 ($\mu$g/$m^3$)', fontdict = {'fontsize':14 });
```

Otro filtro, por ejemplo
¿En qué día se rebasó el nivel recomendado en 2022?

```python
df['PM25']['2022-01-01':'2022-12-31'][df['PM25']>25]
```

Aquí le ponemos una línea para que se vea cuales fueron los valores arriba de 25

```python
fig, ax = plt.subplots(figsize=(10,5),dpi=100)
ax.plot(df['PM25']['2022-01-01':'2022-12-31'], color='gray',lw=0.6);
ax.tick_params(labelsize=14)
ax.set_xlabel('Tiempo', fontdict = {'fontsize':14 })
ax.set_ylabel('PM25 ($\mu$g/$m^3$)', fontdict = {'fontsize':14 });
ax.axhline(25,color='red',lw=2,linestyle='--')
plt.show()
```
