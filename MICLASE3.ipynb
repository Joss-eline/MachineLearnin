```python
import pymannkendall as mk
```

todo sobre el ozono

```python
mk_03 = mk.original_test(df['O3'],alpha=0.05)
mk_03
```

```python
a=mk_03.slope
b=mk_03.intercept
```

hay que definir la funciÃ³n

```python
def f5(x):
  return a*x+b
```

eso me va a regresar los valores en y

```python
x = range(0,3073)
fig, ax = plt.subplots(figsize=(10,5),dpi=200)
ax.plot(x,df['O3'],lw=1,color='cornflowerblue');
ax.plot(x,[f5(i) for i in x],lw=1,color='crimson');

```
